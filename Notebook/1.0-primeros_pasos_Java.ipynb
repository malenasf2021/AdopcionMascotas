
# Primeros Pasos con Java: Instalación y Primer Programa

Antes de comenzar a programar en Java, es necesario instalar el Java Development Kit (JDK). Puedes elegir entre varias versiones y proveedores del JDK, pero las más comunes son el JDK de Oracle y el JDK proporcionado por Eclipse Adoptium.

## Descarga del JDK

### Oracle JDK
El JDK de Oracle es ampliamente utilizado en la industria. Puedes descargar la última versión desde el sitio web oficial de Oracle:

[Descargar Oracle JDK](https://www.oracle.com/java/technologies/javase-jdk15-downloads.html)

### Eclipse Adoptium (AdoptOpenJDK)
Eclipse Adoptium ofrece versiones gratuitas y de código abierto del JDK, compatibles con una amplia gama de sistemas operativos. Puedes descargarlo desde:

[Descargar Eclipse Adoptium (AdoptOpenJDK)](https://adoptium.net/)

## Verificación de la Instalación de Java

Una vez instalado el JDK, verifica que la instalación fue exitosa ejecutando el siguiente comando en una terminal o línea de comandos:

```bash
java -version
```

La salida mostrará la versión instalada de Java, indicando que Java está listo para ser utilizado.

## Tu Primer Programa en Java

A continuación, escribe tu primer programa en Java. Este programa imprimirá "Hello, world!" en la consola.

```java
public class HelloWorld {
    public static void main(String[] args) {
        System.out.println("Hello, world!");
    }
}
```

Guarda este código en un archivo llamado `HelloWorld.java`.

### Compilación y Ejecución

Compila el programa usando el siguiente comando en la terminal:

```bash
javac HelloWorld.java
```

Y ejecútalo con:

```bash
java HelloWorld
```

Si ves "Hello, world!" en la consola, has configurado correctamente tu entorno de desarrollo Java.
