# Guía para Instalar el Kernel de Java en Jupyter Notebook

## Requisitos Previos

### Java JDK

1. Asegúrese de tener instalado **Java JDK 9 o superior**. Puede verificar la versión de Java con el siguiente comando:

    ```bash
    java -version
    ```

    Si no tiene Java instalado, puede instalarlo con:

    ```bash
    sudo apt update
    sudo apt install openjdk-11-jdk
    ```

### Jupyter Notebook

2. Asegúrese de tener instalado **Jupyter Notebook**. Puede verificar la instalación con:

    ```bash
    jupyter --version
    ```

    Si falta algún componente, por ejemplo `jupyter-client`, instálelo con:

    ```bash
    pip install jupyter_client
    ```

    Si no tiene Jupyter Notebook instalado, instálelo con:

    ```bash
    pip install jupyter
    ```

## Instalación del Kernel de Java

### Desde la Fuente

#### Clonar el repositorio

1. Descargue el proyecto desde [el repositorio de GitHub de IJava](https://github.com/SpencerPark/IJava):

    ```bash
    git clone https://github.com/SpencerPark/IJava.git
    cd IJava/
    ```

#### Compilar e instalar

2. Compile e instale el kernel:

    En sistemas Linux:

    ```bash
    ./gradlew installKernel
    ```

    En sistemas Windows:

    ```bash
    gradlew installKernel
    ```
